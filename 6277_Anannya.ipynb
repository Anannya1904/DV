# **Q. What Data Visualization techniques have you used?**

**1) COUNT PLOT**

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=df_data['Continents'])
plt.title("Airlines data Plot")
plt.xlabel("Continents")
plt.ylabel("count")
plt.show()
```

The countplot is used to represent the occurrence(counts) of the observation present in the categorical variable.
It uses the concept of a bar chart for visual depiction. This Count Plot shows that 3000 is the highest count.
Here it counts all regions using Continents variable and **8000 has highest count among the continents**

**2) Hist Plot**

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Continents',color='skyblue',
edgecolor='black',data=df_data)
plt.title("Airlines data Plot")
plt.xlabel('Continents')
plt.ylabel('Count')
plt.show()
```

A histogram is basically used to represent data provided in a form of some groups.It is accurate method for the graphical representation of numerical data distribution.


**3) Scatter Plot**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Continents', y='Age', data=df_data,
color='skyblue')
plt.title("Scatter plot")
plt.show()
```

Scatter plots are used to plot data points on horizontal and
vertical axis in the attempt to show how much one variable is affected by
another

**4) Violin Plot**

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Gender', y='Age', data=df_data)
plt.show()
```

A violin plot combines a box plot with a kernel density estimation,showing the distribution of a variable while also displaying summary statistics. In the above violinplot we can see age group belonging to 20-40 is travelling more in male than female. Whereas age group belonging 45-60 are more in females. The central value i.e. common age group among males and females is 40-60.

**5) Heatmap**

```python
df_data.corr()
```

```python
corr = df_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr,annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

Heatmaps provide us with an easy tool to
understand the correlation between two entities.

**6) Pie Chart**

```python
plt.figure(figsize=(12,8))
df_data['Flight Status'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

It is a circular statistical pie plot that is used to display only one
series of data. The complete area of the pie chart is equal to the total
percentage of the given data. In the Pie Chart, the area of slices of the pie is
used to represent the percentage of the Flight Status. Highest percentage shows the flights that have been cancelled which is 33.4%

**7) Group by Plot**

```python
df_data.groupby('Continents')[['Passenger ID','Age']].sum().plot.bar(color=['blue','red'])
plt.ylabel('Passenger ID and Age')
plt.title('Difference betweeen Passenger ID and Age in Continents')
plt.show()
```

Groupby function is very useful and help full to explain and it
splits the data into groups based on columns/conditions and then apply some
operations. The groupby() can also be applied in series. sort : Sort
group Keys. and that continents has highest passenger ID
